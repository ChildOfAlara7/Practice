```python
countryNames = df.country.unique()
countryNames
```

```python
sums = []
for i in range(len(countryNames)):
    sums.append(df[df['country']== countryNames[i]]['daily_vaccinations'].sum())
```

```python
df2 = pd.DataFrame()
df2['Country'] = countryNames
df2['Sum'] = sums
df2_sorted = df2.sort_values('Sum', ascending=False)
df2_sorted.head(15)
```

```python
top5c = list(df2_sorted['Country'].iloc[0:5])
top5s = list(df2_sorted['Sum'].iloc[0:5])
top5c.append(df2_sorted['Country'].loc[46])
top5s.append(df2_sorted['Sum'].loc[46])
plt.barh(top5c, 
         top5s, 
         height=0.5)
plt.rcParams['figure.figsize'] = [10, 5]
plt.grid()
plt.savefig('graph_top5.png')
```

```python
vacs = []
for i in range(len(countryNames)):
    vacs.append(list(df[df['country']== countryNames[i]]['vaccines'].unique())[0])
```

```python
df2['Vaccine'] = vacs
df2
```

```python
vacNames = list(df.vaccines.unique())
for i in range(len(vacNames)):
    vacNames[i]
#vacNames
```

```python
vac_sums = []
for i in range (len(vacNames)):
    vac_sums.append(df2[df2['Vaccine']== vacNames[i]]['Sum'].sum())
```

```python
df3 = pd.DataFrame()
df3['Vaccine'] = vacNames
df3['vac_sums'] = vac_sums
df3 = df3.sort_values('vac_sums', ascending=False)
df3
```

```python
plt.barh(df3['Vaccine'], 
         df3['vac_sums'], 
         height=0.5)
plt.rcParams['figure.figsize'] = [25, 10]
plt.grid()
plt.savefig('graph_vacs.png')
```
