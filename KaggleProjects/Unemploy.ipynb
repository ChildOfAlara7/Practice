```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('LFSI_UNEMP.csv')
df
```

```python
df.rename(columns={'Unnamed: 0':'country'}, inplace=True )
df = df.drop(['Unnamed: 19'], axis=1)
df = df.drop(11, axis=0)
df = df.fillna(method = 'bfill', axis=1)
```

```python
df
```

```python
Names = list(df['country'].unique())

for i in range(len(df['country'])):
    df = df.rename({i:Names[i]}, axis='index')
    
df = df.drop(['country'], axis=1)
df
```

```python
df = df.T
```

```python
plt.plot(df['Belgium'], 'g', label = 'BE')
plt.plot(df['France'], 'orange', label = 'FR')
plt.plot(df['Netherlands'], 'b', label = "NL")
plt.plot(df['Germany'], 'c', label = "DE")
plt.plot(df['Austria'], 'm', label = "AU")
plt.plot(df['United Kingdom'], 'y', label = "UK")
plt.plot(df['Euro Area 19'], 'k--', label = 'EU19')
plt.plot(df['Europe 28'], 'r--', label = "EU28")
plt.title('Unemployed Level - Europe', size = 18)
plt.grid()
plt.legend(ncol = 3)
plt.rcParams['figure.figsize'] = [12, 6]
plt.savefig('UL_EU.png')
```

```python
plt.plot(df['Belgium'], 'g--', label = 'BE')
plt.plot(df['Brussels-Capital Region'], 'c', label = "BCR")
plt.plot(df['Flemish Region'], 'm', label = "FlR")
plt.plot(df['Walloon Region'], 'y', label = "WR")
plt.title('Unemployed Level - Belgium', size = 18)
plt.grid()
plt.legend(loc='best')
plt.rcParams['figure.figsize'] = [12, 6]
plt.savefig('UL_BE.png')
```
