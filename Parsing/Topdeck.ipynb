```python
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
id = 7296
amount = 10
site = 'https://topdeck.ru/apps/toptrade/member/' + str(id) + '/aucs'
response = r.get(site)
response.status_code
```

```python
soup = bs(response.text, 'lxml')
products = soup.findAll("tr")
del products [0]
```

```python
winners = []
titles = []
link_lot = []
link_win = []
links = []
prices = []
words = []
for p in products:
    pr = str(list(p)[5]).replace('<td data-after="р." data-before="ставка ">\n', '').replace('<td>\n ', '').replace('</td>', '').strip()
    if pr == 'нет ставок': pr = 0
    pr = int(pr)
    prices.append(pr)
    for link in p.find_all('a'):
        links.append(link)
    for word in p.find_all('a'):
        words.append(list(word))
        
for i in range(0, len(words), 2):
    titles.append(str(words[i]).replace("[", ' ').replace("]", ' ').replace("'", ' ').replace('"', ' '))

for i in range(2, len(words)+1, 2):
    winners.append(str(words[i-1]).replace("[", ' ').replace("]", ' ').replace("'", ''))
    
for i in range (0, len(links), 2):
    link_lot.append("topdeck.ru" + links[i].get('href'))
    
for i in range(2, len(words)+1, 2):
    link_win.append("topdeck.ru" + links[i-1].get('href'))
```

```python
df = pd.DataFrame()
df['title'] = titles
df['link_lot'] = link_lot
df['winner'] = winners
df['link_win'] = link_win
df['price'] = prices
```
